```python
import math

def factorial(n):
    return math.factorial(n)

def combinations(n, k):
    return factorial(n) // (factorial(k) * factorial(n - k))

n = 60
k = 6
total_combinations = combinations(n, k)
print(f"O número total de combinações possíveis é: {total_combinations}")
```

```python
import itertools
import csv

# Definindo os valores e o tamanho da combinação
valores = list(range(1, 26))
tamanho_combinacao = 20

# Gerando todas as combinações possíveis
combinacoes = itertools.combinations(valores, tamanho_combinacao)

# Salvando as combinações em um arquivo CSV
with open('combinacoes.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow([f'Número {i+1}' for i in range(tamanho_combinacao)])  # Cabeçalho
    for combinacao in combinacoes:
        writer.writerow(combinacao)

print("Arquivo CSV com todas as combinações gerado com sucesso.")
```

```python
import csv

# Defina a sequência de números a ser pesquisada
concurso = [1, 3, 4, 6, 7, 9, 10, 11, 13, 14, 15, 17, 20, 23, 24 ]

# Inicialize o contador
contador = 0

# Abra o arquivo CSV e leia as combinações
with open('combinacoes.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Pular o cabeçalho
    for row in reader:
        # Converta a linha lida para uma lista de inteiros
        combinacao = list(map(int, row))
        # Verifique se a sequência está na combinação
        if set(concurso).issubset(combinacao):
            contador += 1
            print(f"Combinação vencedora encontrada: {combinacao}")

print(f"A sequência foi encontrada {contador} vezes.")
```

```python
# contar quantos acertos cada combinação de 20 teve

import csv

# Caminhos para os arquivos
combinacoes_file = 'c:/Users/RODRIGO/Desktop/Sena/combinacoes.csv'
resultados_file = 'c:/Users/RODRIGO/Desktop/Sena/resultados_formatados.csv'
output_file = 'c:/Users/RODRIGO/Desktop/Sena/maisCertos.csv'

# Ler as combinações do arquivo combinacoes.csv
combinacoes = []
with open(combinacoes_file, 'r') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Pular o cabeçalho
    for row in reader:
        combinacoes.append(set(map(int, row)))

# Inicializar um dicionário para contar os acertos
acertos_dict = {tuple(combinacao): 0 for combinacao in combinacoes}

# Processar os resultados e contar os acertos
with open(resultados_file, 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        resultado = set(map(int, row))
        for combinacao in combinacoes:
            if resultado.issubset(combinacao):
                acertos_dict[tuple(combinacao)] += 1

# Filtrar as combinações com mais acertos
mais_acertos = {k: v for k, v in acertos_dict.items() if v > 0}

# Escrever os resultados no arquivo maisCertos.csv
with open(output_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Combinação', 'Acertos'])
    for combinacao, acertos in sorted(mais_acertos.items(), key=lambda item: item[1], reverse=True):
        writer.writerow([','.join(map(str, combinacao)), acertos])

print("Arquivo maisCertos.csv gerado com sucesso.")
```

```python
import csv

# Caminhos para os arquivos
mais_certos_file = 'c:/Users/RODRIGO/Desktop/Sena/maisCertos.csv'
resultados_file = 'c:/Users/RODRIGO/Desktop/Sena/resultados_formatados.csv'
output_file = 'c:/Users/RODRIGO/Desktop/Sena/maisCertos_nova.csv'

# Ler as combinações e seus acertos do arquivo maisCertos.csv
combinacoes_acertos = []
with open(mais_certos_file, 'r') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Pular o cabeçalho
    for row in reader:
        combinacao = set(map(int, row[0].split(',')))
        acertos = int(row[1])
        combinacoes_acertos.append([combinacao, acertos])

# Ler as linhas do arquivo resultados_formatados.csv
sorteios = []
with open(resultados_file, 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        sorteios.append(set(map(int, row)))

# Processar as combinações e remover as correspondentes dos sorteios
for sorteio in sorteios[:]:
    for combinacao_acertos in combinacoes_acertos:
        combinacao, acertos = combinacao_acertos
        if sorteio.issubset(combinacao):
            combinacao_acertos[1] += 1
            sorteios.remove(sorteio)
            break

# Filtrar as combinações com mais acertos
mais_acertos_nova = [(combinacao, acertos) for combinacao, acertos in combinacoes_acertos if acertos > 0]

# Escrever os resultados no arquivo maisCertos_nova.csv
with open(output_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Combinação', 'Acertos'])
    for combinacao, acertos in sorted(mais_acertos_nova, key=lambda item: item[1], reverse=True):
        writer.writerow([','.join(map(str, combinacao)), acertos])

print("Arquivo maisCertos_nova.csv gerado com sucesso.")
```

```python
# cria um arquivo com as combinações que tiveram mais acertos únicos chamado maisCertos_nova_atualizado.csv

import csv

# Caminhos para os arquivos
mais_certos_nova_file = 'c:/Users/RODRIGO/Desktop/Sena/maisCertos_nova.csv'
resultados_file = 'c:/Users/RODRIGO/Desktop/Sena/resultados_formatados.csv'
output_file = 'c:/Users/RODRIGO/Desktop/Sena/maisCertos_nova_atualizado.csv'

# Ler as combinações do arquivo maisCertos_nova.csv
jogos = []
with open(mais_certos_nova_file, 'r') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Pular o cabeçalho
    for row in reader:
        combinacao = set(map(int, row[0].split(',')))
        jogos.append([combinacao, 0])  # Inicializar contador de acertos com 0

# Ler as linhas do arquivo resultados_formatados.csv
sorteios = []
with open(resultados_file, 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        sorteios.append(set(map(int, row)))

# Processar os jogos e remover as correspondentes dos sorteios
for jogo in jogos:
    combinacao, acertos = jogo
    for sorteio in sorteios[:]:
        if sorteio.issubset(combinacao):
            jogo[1] += 1
            sorteios.remove(sorteio)

# Filtrar as combinações com mais acertos
mais_acertos_nova = [(combinacao, acertos) for combinacao, acertos in jogos if acertos > 0]

# Escrever os resultados no arquivo maisCertos_nova_atualizado.csv
with open(output_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Combinação', 'Acertos'])
    for combinacao, acertos in sorted(mais_acertos_nova, key=lambda item: item[1], reverse=True):
        writer.writerow([','.join(map(str, combinacao)), acertos])

print("Arquivo maisCertos_nova_atualizado.csv gerado com sucesso.")
```

```python
import csv

# Caminho para o arquivo maisCertos.csv
mais_certos_file = 'c:/Users/RODRIGO/Desktop/Sena/maisCertos_nova_atualizado.csv'

# Ler as combinações e seus acertos do arquivo maisCertos.csv
combinacoes_acertos = []
with open(mais_certos_file, 'r') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Pular o cabeçalho
    for row in reader:
        combinacao = row[0]
        acertos = int(row[1])
        combinacoes_acertos.append((combinacao, acertos))

# Função para calcular o total de acertos para as primeiras N combinações
def calcular_acertos(n):
    total_acertos = sum(acertos for _, acertos in combinacoes_acertos[:n])
    return total_acertos

# Calcular os acertos para as 5, 10, 15 e 20 primeiras combinações
acertos_5 = calcular_acertos(5)
acertos_10 = calcular_acertos(10)
acertos_15 = calcular_acertos(15)
acertos_20 = calcular_acertos(20)

# Se estiver tudo certo, o total de acertos deve ser 3285 em 592 combinações
teste = calcular_acertos(592)



print(f"Total de acertos usando as 5 primeiras combinações: {acertos_5}")
print(f"Total de acertos usando as 10 primeiras combinações: {acertos_10}")
print(f"Total de acertos usando as 15 primeiras combinações: {acertos_15}")
print(f"Total de acertos usando as 20 primeiras combinações: {acertos_20}")

print(f"Total de acertos usando as 592 primeiras combinações: {teste}")



```

```python
# gráfico de quantos acertos por somatório de linhas
import csv
import matplotlib.pyplot as plt

# Caminho para o arquivo maisCertos.csv
mais_certos_file = 'c:/Users/RODRIGO/Desktop/Sena/maisCertos_nova_atualizado.csv'

# Ler as combinações e seus acertos do arquivo maisCertos.csv
combinacoes_acertos = []
with open(mais_certos_file, 'r') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Pular o cabeçalho
    for row in reader:
        combinacao = row[0]
        acertos = int(row[1])
        combinacoes_acertos.append((combinacao, acertos))

# Função para calcular o total de acertos para as primeiras N combinações
def calcular_acertos(n):
    total_acertos = sum(acertos for _, acertos in combinacoes_acertos[:n])
    return total_acertos

# Calcular os acertos para diferentes números de primeiras combinações
x_values = list(range(1, 36))  # De 1 a 36, a partir daqui é prejú
y_values = [calcular_acertos(n) for n in x_values]

# Criar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, marker='o')
plt.title('Número de Acertos vs. Número de Primeiras Combinações')
plt.xlabel('Número de Primeiras Combinações')
plt.ylabel('Número de Acertos')
plt.grid(True)
plt.xticks(x_values)  # Mostrar todos os valores de x
plt.show()
```
